Testing one of the hypotheses:

Average user ratings for the Action and Sports genres are different.

H0: m1 = m2 (the mean user rating for Action is equal to the mean user rating for Sports)

H1: m1 <> m2 (the mean user rating for Action is NOT equal to the mean user rating for Sports)

alpha = 0.05

```python
df.groupby('genre').user_score.describe().loc[['Action', 'Sports']]
```

```python
stats.ttest_ind(df[df.genre=='Action'].user_score.dropna(), df[df.genre=='Sports'].user_score.dropna())
```

We get a VERY small pvalue (1.53x10-6) -- so with alpha set to 0.05 we reject the null hypothesis -- i.e. the two means are statistically different. The average user ratings for the Action and Sports genres are not the same.

if p < alpha ==> reject the null hypotheses.



Testing the other Hypotheses

Average user ratings of the Xbox One and PC platforms are the same.

H0: m1 = m2 (the mean user rating for Xbox One is equal to the mean user rating for PC.)

H1: m1 <> m2 (the mean user rating for Xbox One is NOT equal to the mean user rating for PC)

I chose alpha = 0.05 as the significant level.

```python
df.groupby('platform').user_score.describe().loc[['XOne', 'PC']]
```

```python
stats.ttest_ind(df[df.platform=='XOne'].user_score.dropna(), df[df.platform=='PC'].user_score.dropna())
```

Comments:The pvalue is smaller than alpha which is set to 0.05 so we reject the null hypothesis. So the Average user ratings of the Xbox One and PC platforms are not the same.

I chose this hypotheses because all the tests are independent. There is n cmparison of prior results of the same.
