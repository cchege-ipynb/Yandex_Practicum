How the concentrations of metals (Au, Ag, Pb) change depending on the purification stage.

Building the Model

Using Linear Regression.

```python
target = ['final.output.recovery', 'rougher.output.recovery']
features = train_test_columns[1:]
```

```python
X_train, X_test = df0[features].values, df1[features].values
y_train, y_test = df0[target].values, df1[target].values
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
def smape(y_true, y_pred):
    return (np.abs(y_true - y_pred)/((np.abs(y_true)+np.abs(y_pred))/2)).mean()
```

Defining the SMAPE function and using it.

```python
SMAPE_linear = smape(y_test[:,0], lr.predict(X_test)[:,0])
print('The smape_final is = '+"{:.4}".format(SMAPE_linear))
```

```python
def smape_final(y_true,y_pred):
    smape_out_rougher = smape(y_true[:,1],y_pred[:,1])
    smape_out_final =smape(y_true[:,0],y_pred[:,0])
    return 0.25*smape_out_rougher + 0.75*smape_out_final
```

```python
SMAPE = smape_final(y_test,lr.predict(X_test))
print('The smape_final is = '+"{:.4}".format(SMAPE))
```

Cross Validation - Linear Regression

```python
smape_score = make_scorer(smape_final)
```

```python
scores_lr = cross_val_score(lr, X_train, y_train, cv=5, scoring=smape_score)
print('mean smape:= '+"{:.4}".format(scores_lr.mean()))
scores_lr
```

Checking for the accuracy of the model using the cross-validation score. 0.10 is good enough.

Other Models - DecisionTree Regressor and RandomForest Regressor

```python
for i in [1,2,5,10]:
    dtr = DecisionTreeRegressor(max_depth=i)
    dtr.fit(X_train,y_train) 
    scores_dt = cross_val_score(dtr, X_train, y_train, cv=5, scoring=smape_score)
    print('Max Depth ' + str(i)+', mean sMAPE: ' + str((scores_dt.mean()).round(4)) )
```

```python
for i in range(50,200,50):
    rf = RandomForestRegressor(n_estimators=i)
    rf.fit(X_train,y_train) 
    scores_dt = cross_val_score(rf, X_train, y_train, cv=5, scoring=smape_score)
    print('n_estimators ' + str(i)+', mean sMAPE: ' + str((scores_dt.mean()).round(4)) )
```
