Splitting the data into train, valid and test sets in the ratio 3.1.1.

```python
# Split target and features
target = df['is_ultra']
features = df.drop('is_ultra', axis = 1)

# Split into sets
features_train, features_temp, target_train, target_temp = train_test_split(features, target, test_size=0.2,random_state=42)
features_valid, features_test, target_valid, target_test = train_test_split(features_temp, target_temp, test_size=0.2,random_state=42)
```

Defining the features ie the column variables to be used and targets ie the feature used to make predictions, for the 3 datasets.

```python
for estim in range(1,5,1):
        model = DecisionTreeClassifier(random_state=54321)

        model.fit(features_train,target_train)
        predictions_valid = model.predict(features_valid)
        accuracy = accuracy_score(target_valid, predictions_valid)
        
        model.score(features_train, target_train)
        
        print("n_estimators =", estim, ":", accuracy)
```
